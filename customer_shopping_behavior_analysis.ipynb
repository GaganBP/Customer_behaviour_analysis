Here we are checking for missing values

```python
df.isnull().sum()
```

we are impute the missing values using median review rating within each category

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns=df.columns.str.replace(' ', '_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

here we are creating 4 columns for customer age groups

```python
# Define age group labels in order from youngest to oldest
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']

# Split 'age' column into 4 equal-sized quantile buckets and assign labels
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create a column purchase_frequency_days because it was in text so we do it in numbers
# Dictionary mapping purchase frequency labels to number of days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

# Map frequency labels to numeric days and store in new column
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
# Check if discount_applied and promo_code_used are identical across all rows
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Drop promo_code_used as it is identical to discount_applied (redundant column)
df = df.drop(columns=['promo_code_used'], axis=1)
```
